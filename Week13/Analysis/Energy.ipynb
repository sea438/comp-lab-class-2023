```python
import numpy as np
import matplotlib.pyplot as plt

def read_ener_file(file_path):
    # Read the .ener file and extract relevant columns
    data = np.loadtxt(file_path, comments=['#', '&'])
    time = data[:, 0]
    potential_energy = data[:, 4]
    
    return time, potential_energy

# Read data from the .ener files
time1, energy1 = read_ener_file('../Outputs/NPT-1.ener')
time2, energy2 = read_ener_file('../Outputs/NVT-1.ener')
time3, energy3 = read_ener_file('../Outputs/monitor/MONITOR-1.ener')

# Plot potential energy versus time
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy')

plt.plot(time1, energy1, label='NPT')
plt.title('NPT Potential Energy versus Time')
plt.savefig('../Figures/NPT_potential.png')
plt.show()

plt.plot(time2, energy2, label='NVT')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy')
plt.title('NVT Potential Energy versus Time')
plt.savefig('../Figures/NVT_potential.png')
plt.show()

plt.plot(time3, energy3, label='QM/MM')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy')
plt.title('QM/MM Potential Energy versus Time')
plt.savefig('../Figures/QM_potential.png')
plt.show()

```
