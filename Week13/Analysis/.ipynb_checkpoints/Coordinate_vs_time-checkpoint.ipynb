```python
import numpy as np
import matplotlib.pyplot as plt

colvar_file = ['../Outputs/monitor/MONITOR-COLVAR.metadynLog',
               '../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog',
              '../Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog']
 
for item in colvar_file:
    colvar_data = np.loadtxt(item)

    # Extract time and reaction coordinate
    time = colvar_data[:, 0]
    reaction_coordinate = colvar_data[:, 1]  

    # Plot reaction coordinate versus time
    if item=='../Outputs/monitor/MONITOR-COLVAR.metadynLog':
        plt.plot(time, reaction_coordinate, label='Standard QM/MM')
    elif item=='../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog':
        plt.plot(time,reaction_coordinate, label='Frequency=100,Width=0.5, Height=1.5')
#Frequency change 
    else:
        plt.plot(time,reaction_coordinate, label='Frequency=20,Width=0.5, Height=1.5')
        
    # Customize the plot (e.g., add labels, title, legend, etc.)
    plt.xlabel('Time (ps)')
    #Colvar is 'distance between C-O bond that will be broken and C-C bond that is formed'
    plt.ylabel('Reaction Coordinate (r(CO)-r(CC))')
    plt.title('Reaction Coordinate versus Time')

    # Show the plot
plt.legend()
plt.savefig('../Figures/coordinate_vs_time.png')
plt.show()
```
