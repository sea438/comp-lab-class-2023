```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = '1hz3_T310.run.25000000.energy.xvg'
```

```python
data = np.genfromtxt(filepath, skip_header=29)

time = data[:,0]
potential = data[:,1]
kinetic_e = data[:,2]
total_e = data[:,3]
temp = data[:,4]
pres = data[:,5]
vol = data[:,6]

y_labels={'Potential Energy (kJ/mol)': potential,
         'Kinetic Energy (kJ/mol)': kinetic_e,
         'Total Energy (kJ/mol)': total_e,
         'Temperature (K)': temp,
         'Pressure (bar)': pres,
         'Volume (n^3)':vol}

for i in y_labels:
    plt.figure(figsize=(3,3))
    plt.plot(time,y_labels[i])
    plt.ylabel(i)
    plt.xlabel('Time (ns)')
    plt.title(i + ' vs. Time (ns)')
    plt.show()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(time, temp, label='Temperature (K)', color='blue')
ax1.plot(time,kinetic_e,label='Kinetic Energy (kJ/mol)',color='red')

ax2=ax1.twinx()
ax1.set_ylabel('Temprature (K)',color='blue')
ax2.set_ylabel('Kinetic Energy (kJ/mol)',color='red')
ax1.legend(loc='upper right')

ax1.set_xlabel('Time (ns)')

plt.show()
```
