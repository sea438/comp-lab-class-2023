```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
filepaths = ['T300/energy.xvg','T363/energy.xvg','T440/energy.xvg']

for file in filepaths:
    with open(file, 'r') as f:
        E=np.genfromtxt(f,skip_header=24)
        E=np.ndarray.transpose(E)
    plt.hist(E[1],bins=100,label=file)
    plt.xlabel('Energy')
    plt.ylabel('Frequency')
    plt.legend()
    
with open('T300/energy.xvg','r') as f:
    E=np.genfromtxt(f,skip_header=24)
E=np.ndarray.transpose(E)
fig1,ax1=plt.subplots()
hist_values = ax1.hist(E[1],label='Energy at 300 K')
ax1.set_ylabel('Energy')
ax1.set_xlabel('Probability')
ax1.legend()
```

```python
#Replicas in time
with open('T300/replica_temp.xvg','r') as f:
    data=np.genfromtxt(f)
     
for i in range(0,3):
    replica =[]
    time=[]
    for t in range(0,500,20):
        time += [data[t,0]]
        replica+=[data[t,i]]
    plt.title('Replica ' + str(i+1) +' vs. time')
    plt.xlabel('Time')
    plt.ylabel('Replica Temp')
    plt.plot(time,replica)
plt.show()
```
