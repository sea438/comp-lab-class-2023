```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading in COLVAR file 
filepath = 'Outputs/lugano1d_phi/ala2_metad_biasfactor10_pace500_height1.2_sigma0.55_10ns.colvar.txt'
data = np.genfromtxt(filepath,skip_header=9)

#Plotting phi and psi from COLVAR
time=data[:,0]
phi=data[:,1]

#Plottingonly every 50th point because otherwise the plot 
#looks like a big blob (points too close together)

time_subset=time[::50]
phi_subset=phi[::50]


plt.scatter(time_subset,phi_subset)
plt.xlabel('Time')
plt.ylabel('Phi')
plt.title('Phi vs. Time')
plt.savefig('Figures/Phi vs Time Lugano Tutorial 1_1d_phipsi.png')
```

```python
#Loading in fes file 
filepath = 'Outputs/lugano1d_phi/fes.dat'
data = np.genfromtxt(filepath,skip_header=9)

#Plotting phi and psi versus energy
phi=data[:,0]
energy=data[:,1]

fig1, ax1=plt.subplots()
ax1.plot(phi,energy)
ax1.set_title('Energy vs. Phi')
ax1.set_ylabel('Energy')
ax1.set_xlabel('Phi')
fig1.savefig('Figures/Energy vs. Phi Lugano Tutorial 1 1D.png')
```
