```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
def analyze(file):
    #Loading in COLVAR file 
    filepath = file
    data = np.genfromtxt(filepath,skip_header=5)

    #Plotting phi and psi from COLVAR
    time=data[:,0]
    phi=data[:,1]
    psi=data[:,2]

    fig1, ax1=plt.subplots()
    ax1.plot(time,phi)
    ax1.set_title('Progression of Phi in Time')
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Phi')
    fig1.savefig('Figures/Phi vs. Time.png')

    fig2, ax2=plt.subplots()
    ax2.plot(time,psi)
    ax2.set_title('Progression of Psi in Time')
    ax2.set_xlabel('Time')
    ax2.set_ylabel('Psi')
    fig2.savefig('Figures/Psi vs. Time.png')

    fig3, ax3 =plt.subplots()
    ax3.scatter(phi,psi)
    ax3.set_xlabel('Phi')
    ax3.set_ylabel('Psi')
    fig3.savefig('Figures/Psi vs. Phi.png')
    
    #Standard dev calculations
    sigma_phi=np.std(phi)
    print(sigma_phi)

    sigma_psi=np.std(psi)
    print(sigma_psi)
    print()
    
analyze('Outputs/A/COLVAR')
analyze('Outputs/B/COLVAR')
```

```python
#Loading in phi and psi from xtc files

traj = md.load('Outputs/A/traj_comp.xtc', top='Outputs/A/confout.gro')

#Calculating phi and psi using mdtraj
phi_indices, phi_angles = md.compute_phi(traj)
psi_indices, psi_angles = md.compute_psi(traj)
time=traj.time

fig4, ax4 =plt.subplots()
ax4.scatter(phi_angles,psi_angles)
ax4.set_xlabel('Phi')
ax4.set_ylabel('Psi')
fig4.savefig('Figures/Psi vs. Phi from xtc file A.png')

###YAY THESE MATCH!

#Repeating for B

traj = md.load('Outputs/B/traj_comp.xtc', top='Outputs/B/confout.gro')

#Calculating phi and psi using mdtraj
phi_indices, phi_angles = md.compute_phi(traj)
psi_indices, psi_angles = md.compute_psi(traj)
time=traj.time

fig5, ax5 =plt.subplots()
ax5.scatter(phi_angles,psi_angles)
ax5.set_xlabel('Phi')
ax5.set_ylabel('Psi')
fig5.savefig('Figures/Psi vs. Phi from xtc file B.png')

#They match times two :)
```
