```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading in COLVAR file 
filepath = 'Outputs/lugano2d_phipsi/ala2_metad_biasfactor10_pace500_height1.2_sigma0.55_1.02_10ns.colvar.txt'
data = np.genfromtxt(filepath,skip_header=9)

#Plotting phi and psi from COLVAR
time=data[:,0]
phi=data[:,1]
psi=data[:,2]

#Plottingonly every 50th point because otherwise the plot 
#looks like a big blob (points too close together)
time_subset=time[::50]
phi_subset=phi[::50]
psi_subset=psi[::50]

fig, ax=plt.subplots()
ax.scatter(time_subset,phi_subset)
ax.set_xlabel('Time')
ax.set_ylabel('Phi')
fig.savefig('Figures/Phi vs Time Lugano Tutorial 1_2d_phipsi.png')

fig1, ax1=plt.subplots()
ax1.scatter(time_subset,psi_subset)
ax1.set_xlabel('Time')
ax1.set_ylabel('Psi')
fig.savefig('Figures/Psi vs Time Lugano Tutorial 1_2d_phipsi.png')
```

```python
#Loading in grid file 
filepath = 'Outputs/lugano2d_phipsi/gridfile.grid.dat'
data = np.genfromtxt(filepath,skip_header=9)

#Plotting phi and psi from COLVAR
phi=data[:,0]
psi=data[:,1]

der_phi=data[:,3]
der_psi=data[:,4]
free_e=[der_phi,der_psi]
free_e=np.transpose(free_e)

PHI, PSI = np.meshgrid(phi,psi)

fig, ax = plt.subplots()
ax.set_xlabel('Phi')
ax.set_ylabel('Psi')
ax.set_title('Free Energy')

# Display the 2D hills using imshow
cmap = plt.get_cmap('viridis')  # You can choose a different colormap
im = ax.imshow(free_e, extent=(phi.min(), phi.max(), psi.min(), psi.max()), cmap=cmap, origin='lower')
# Add a colorbar to indicate energy values
cbar = fig.colorbar(im, ax=ax)

plt.savefig('Figures/Energy Lugano Tutorial 1_2d_phipsi.png')
```
