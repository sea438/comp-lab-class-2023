```python
traj = md.load_xtc('Setup/gromacs/protein_only_traj.xtc',
top = '2jof.pdb')

def end_end(traj):
    """Compute the end to end distance of a trajectory"""
    # get the positions of the first and last atoms
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atoms
    dist = np.linalg.norm(first - last, axis=1)
    return dist

# compute the end to end distance
dist = end_end(traj)

plt.plot(time,dist)
plt.ylabel("End to end distance(nm)")
plt.xlabel('Time (ps)')
plt.title('End to End Distance vs. time')
plt.show()

plt.savefig('/scratch/work/courses/CHEM-GA-2671-2023fa/students/sea438/comp-lab-class-2023/Week4/Figures/end_end.png')
```
