```python
###Looking at this it looks like I missed a few zeros when 
###trying to run 50 ns simulation. Pushing this for now until run a 50 ns one

filepath = 'rmsd50.xvg'
data = np.genfromtxt(filepath,skip_header=18)

filepath2='rmsd_50_xtal.xvg'
data2 = np.genfromtxt(filepath2,skip_header=18)

time = data[:,0]
rmsd50 = data[:,1]

time2 = data2[:,0]
rmsd_50_xtal = data2[:,1]

plt.plot(time,rmsd50)
plt.plot(time2,rmsd_50_xtal)
plt.ylabel("RMSD (nm)")
plt.xlabel('Time (ns)')
plt.title('RMSD Backbone 50 ns Simulation')
plt.legend(labels=['Ref:Equilibrated','Ref:Crystal Structure'])
plt.show()
```
