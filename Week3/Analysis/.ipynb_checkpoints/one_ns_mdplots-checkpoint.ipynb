```python
filepath = 'rmsd.xvg'
data = np.genfromtxt(filepath,skip_header=18)

filepath2='rmsd_xtal.xvg'
data2 = np.genfromtxt(filepath2,skip_header=18)

time = data[:,0]
rmsd = data[:,1]

time2 = data2[:,0]
rmsd_xtal = data2[:,1]

plt.plot(time,rmsd)
plt.plot(time2,rmsd_xtal)
plt.ylabel("RMSD (nm)")
plt.xlabel('Time (ns)')
plt.title('RMSD Backbone')
plt.legend(labels=['Ref:Equilibrated','Ref:Crystal Structure'])
plt.show()
```

```python
#Radius of gyration
filepath = 'gyrate.xvg'
data = np.genfromtxt(filepath,skip_header=27)

time = data[:,0]
r_gyration = data[:,1]

plt.plot(time,r_gyration)
plt.ylabel("Radius of gyration (nm)")
plt.ylim(1.30,1.50)
plt.xlabel('Time (ns)')
plt.title('Radius of Gyration (Unrestrained MD)')
plt.show()
```
