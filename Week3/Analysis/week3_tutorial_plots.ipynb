```python
#Energy Minimization

filepath = 'potential.xvg'
data = np.genfromtxt(filepath,skip_header=24)

energy_min_step = data[:,0]
potential = data[:,1]

plt.plot(energy_min_step,potential)
plt.ylabel("Potential Energy (kJ/mol)")
plt.xlabel('Energy Minimization Step')
plt.title('Eenrgy Minimization')
plt.show()

#Saving into Figures folder

plt.savefig('/scratch/work/courses/CHEM-GA-2671-2023fa/students/sea438/comp-lab-class-2023/Week3/Figures/energy_min_plot.png')
```

```python
#Importing data from md simulation
filepath = 'temperature.xvg'
data = np.genfromtxt(filepath,skip_header=24)

#Reading data into python
time = data[:,0]
temperature = data[:,1]
```
